## Start the analysis

### Task 1

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Import lego_sets and quickly review the dataset
lego_sets = pd.read_csv('datasets/lego_sets.csv')
lego_sets.head()
```

```python
# Import parent_themes and quickly review the dataset
parent_themes = pd.read_csv('datasets/parent_themes.csv')
parent_themes.head()
```

```python
# Check the missing value of lego_sets
print(lego_sets.info())

lego_miss = lego_sets.isnull().sum()
lego_miss.plot.bar()
```

```python
# Check the missing value of _sets
parent_themes.info()
```

```python
# Exclude the num_parts and dropna of set_num column
lego_sets.drop('num_parts', axis=1, inplace=True)
lego_sets.dropna(subset=['set_num'], inplace=True)
print(lego_sets.info())
lego_sets.head()
```

```python
# merge the lego_sets and parent_themes
data = pd.merge(lego_sets, parent_themes, 
                left_on='parent_theme', right_on='name', 
                how='left', suffixes=['_set', '_theme'])
print('The DataFrame has {} rows and {} columns'.format(data.shape[0], data.shape[1]))
data.head()
```

```python
# Calucate the result of task 1
num_of_star = data[(data['is_licensed'] == True) & (data['name_theme'] == 'Star Wars')].shape[0]
num_of_licensed = data[data['is_licensed'] == True].shape[0]

# the force
the_force = round(num_of_star / num_of_licensed * 100)
```

### Task 2

```python
# Count the number of theme for different years
licensed_df = data[data['is_licensed'] == True]

group_result = licensed_df.groupby(['year', 'name_theme']).agg({'name_theme': 'count'})
group_result.rename(columns={'name_theme': 'number'}, inplace=True)
group_result.reset_index(inplace=True)

# The the first few rows for the dataframe
group_result.head()
```

```python
# Copy the group_result table and reset_index
group_max = group_result.copy().reset_index()
group_max = group_max.groupby('year').agg({'number': 'max'})
group_max = group_max.reset_index()

# Quickly check the result
group_max.head()
```

```python
# Merge two tables
result_df = pd.merge(group_max, group_result, on=['year', 'number'], how='left')
result_df.head()
```

```python
# Check which year the most popular theme is not star war
new_era = int(result_df[result_df['name_theme'] != 'Star Wars']['year'])
print(new_era)
```
